```python
import pandas
import numpy as np
```

```python
#open data/meteo_m.txt as csv in Pandas
df = pandas.read_csv('data/meteo_m.txt', sep=';', names=["DateTime","WS","WD","Temp","RelHum","Press","PrecType","PrecIntens","PrecAccum","SolRad","SolAccum","UV","AvgWS_10min","MaxWS_10min","MinWS_10min","AvgDir_10min","MaxDir_10min","MinDir_10min","WGust","AvgWS_2min","MaxWS_2min","MinWS_2min","AvgWD_2min","MaxWD_2min","MinWD_2min"])
df["DateTime"] = pandas.to_datetime(df["DateTime"],format='%Y-%m-%d %H:%M:%S ')
```

```python
def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))
x = np.arange(len(df))
fit = 580*gaussian(x, 318, 150) + 580*gaussian(x, 1750, 150)

#add a new column gauss to the dataframe with g as values
df['SolRadFit'] = fit

#plot "Gauss" and "SolRad" column
df.plot(x='DateTime', y=['SolRad', 'SolRadFit'])

```

```python
#plot Temp
df.plot(x="DateTime", y="Temp", color="red")
#plot RelHum
df.plot(x="DateTime", y="RelHum", color="blue")
#plot Press
df.plot(x="DateTime", y="Press", color="green")
```
