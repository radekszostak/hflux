```python
import pandas
```

```python
#open doplyw.csv and most.csv in pandas
doplyw = pandas.read_csv('data/doplyw.csv', names=["date", "time", "temperature"])
most = pandas.read_csv('data/most.csv', names=["date", "time", "temperature"])
#concatenate two first columns in dataframe
doplyw['date_time'] = doplyw['date'] + ' ' + doplyw['time']
most['date_time'] = most['date'] + ' ' + most['time']
#parse date_time string to datetime
doplyw['date_time'] = pandas.to_datetime(doplyw['date_time'], format='%d/%m/%Y %H:%M:%S')
most['date_time'] = pandas.to_datetime(most['date_time'], format='%d/%m/%Y %H:%M:%S')
#remove columns "date" and "time"
doplyw.drop(['date', 'time'], axis=1, inplace=True)
most.drop(['date', 'time'], axis=1, inplace=True)
#swap columns "date_time" and "temperature"
doplyw = doplyw[['date_time', 'temperature']]
most = most[['date_time', 'temperature']]
```

```python
#plot temperature in doplyw and most in one figure
import matplotlib.pyplot as plt
plt.plot(most['date_time'], most['temperature'], label='most')
plt.plot(doplyw['date_time'], doplyw['temperature'], label='doplyw')

#plot legend
plt.legend()
```

```python
#subract temperature from most from doplyw
(doplyw['temperature'] - most['temperature']).plot()
```
